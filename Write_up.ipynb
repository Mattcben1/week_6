## First Steps – Data Organization and Clustering

I took the datasets I was provided and ensured they were clean before merging them. This process is necessary and ensures that we remove all missing entries. After merging the two datasets, I moved on to the unsupervised machine learning algorithm known as K-means Clustering. Essentially, this algorithm creates clusters based on how many data points are closest to the centroid (center). In essence, we use it to identify or group similar players based on the features that best represent their skill level.

I chose Points, Total Rebounds, Assists, Steals, Blocks, Turnovers, Games Started, and Minutes Played. To finalize creating our clustering algorithm, we scale the features we are using to make them equally important. For example, the value of 2 rebounds is different than 10 points. Wrapping up our K-means algorithm, we call the silhouette score function, which associates a score based on how well the data is being clustered. The score of my first algorithm with four clusters was 41% on a 100-point scale. However, in basketball and many other sports, skill features are effects of others, meaning that the clusters are most likely to overlap, as a player usually has many specialties.

## Observation – A Different Approach To The Algorithm

After evaluating the clustering results above, I began to wonder if this algorithm was truly aligning with the objective. Since the current dataset is combined season stats, I was curious about the risk of missing undervalued players. For example, players who may have been injured and missed games, underused players who are solid contributors, and breakout players who aren’t lead scorers but show a positive trend.

Seasonal stats caused these types of player cases to be flagged as unproductive in the clustering algorithm, which we know is not entirely true. To combat this, I converted the season stats into per-game stats by taking the original features mentioned earlier and dividing them by the player-specific Games Played statistic. I then filtered the Minutes Played statistic to greater than or equal to 12, ensuring that we exclude predominantly bench players. This left us with 435 entries.

## Results – Filtering and Analyzing Salary vs. Skill

Now that we have our filtered parameters in place, we can analyze which groups of players are honing their skills. For example, in scoring vs. rebounding, shooting guards seem to be at the middle to bottom right of the plot, while centers place more near the top, or blue cluster (1).

To decide which players seem like a good “deal,” we need to look further into the players' salaries and how they can be combined with their skill level. By creating a “Cluster Average Salary” variable, we pull the mean salary of players within a certain cluster. Furthermore, when we subtract that variable from the player's salary, we receive our verdict. If the outcome of that function is negative for a specific player, then they are of similar skill value to a player whose salary is positive. Theoretically, I think this would be beneficial with an increased number of clusters, but once you’re past a certain point, the threshold between player skill and position starts to blur.
